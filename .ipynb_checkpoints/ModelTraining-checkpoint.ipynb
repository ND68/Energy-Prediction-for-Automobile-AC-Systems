```python
df = pd.read_csv('./Data/Merged/api_merged.csv')
pd.set_option('display.max_columns', None)
df.insert(0, 'Aircon Consumption Power (kW)', df.pop('Aircon Consumption Power (kW)')) # moves A/C power consumption to front
```

## Data Normalization (Scaling)
**Z-score normalization**: Handles outliers, but does not produce normalized data with the *exact* same scale.

```python
#Z-score normalization (standardization)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
standard_scaled_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns) 
```

```python
standard_scaled_df.describe()
```

**Min-max normalization**: Guarantees all features will have the exact same scale but does not handle outliers well.

```python
#Min-Max normalization
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
min_max_scaled_df = pd.DataFrame(min_max.fit_transform(df), columns = df.columns)
```

```python
min_max_scaled_df.describe()
```

## Data Splitting

```python
scaled_df = min_max_scaled_df
x = scaled_df.copy()
y = x.pop('Aircon Consumption Power (kW)')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, shuffle=False)
```

## Feature Selection

**Principle Component Analysis (PCA)**: a dimensionality reduction method; transforms a large set of variables into a smaller one

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1) #turns all x values into 1
pca_x_train = pca.fit_transform(x_train)
```

```python
plt.scatter(pca_x_train, y_train, color='#947698')
plt.show()
```

**Mutual Information**: indicates how much information can be obtained from a random variable by observing another random variable. 
- High mutual information value = a large reduction of uncertainty (knowing x will tell you something about y)
- Low value = a small reduction 
- If the mutual information is zero, the two variables are independent (knowing x won't tell you anything about y)

```python
from sklearn.feature_selection import mutual_info_regression
mutual_info = mutual_info_regression(x,y)
mutual_info = pd.Series(mutual_info, x.columns)
```

```python
mutual_info.sort_values(ascending=True, inplace=True)
f = plt.figure(figsize=(8,8))
ax = f.add_subplot(111)
ax.yaxis.tick_right()
ax.tick_params(axis="y",direction="in", pad=-5)
plt.title("Mutual Information Scores (Top Features)", fontsize=10)
mutual_info[len(mutual_info)-15:].plot(kind='barh', color='#00BB2D')
plt.grid(True, axis='x', alpha=0.3)
plt.xlim(2.5,0)
plt.yticks(fontsize=10, ha='right')
plt.xticks(fontsize = 8)

#plt.savefig("mutual_info.png", format='png', bbox_inches='tight')
plt.show()
```

**Variance Threshold**: Identifies columns that don't vary much in terms of data collected. Low variance means that its values are mostly constant or do not change much across the dataset.

```python
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=0)
vt.fit(x)
variances = vt.variances_
variances = pd.Series(variances, x.columns)
```

```python
variances.sort_values(ascending=True, inplace=True)
plt.figure(figsize=(10,10))
plt.title("Feature Variances")
variances.plot(kind='barh', color='#6D3F5B')
plt.grid(True, axis='x', alpha=0.3)
plt.show()
```

```python
def build_and_compile_model(norm):
  model = keras.Sequential([
      norm,
      layers.Dense(64, activation='relu'),
      layers.Dense(64, activation='relu'),
      layers.Dense(1)
  ])

  model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
  return model
```

### Stacked LSTM sequence to sequence Autoencoder

```python
train_df = scaled_df[:round(len(scaled_df)*.8)]
test_df = scaled_df[round(len(scaled_df)*.8):]
```

```python
# sliding window

def split_series(series, n_past, n_future):
  
  # n_past ==> no of past observations
  # n_future ==> no of future observations 

  l1, l2 = list(), list()
  for window_start in range(len(series)):
    past_end = window_start + n_past
    future_end = past_end + n_future
    if future_end > len(series):
      break
    # slicing the past and future parts of the window
    past, future = series[window_start:past_end, :], series[past_end:future_end, :]
    l1.append(past)
    l2.append(future)
  return np.array(l1), np.array(l2)

n_past = 10
n_future = 5
n_features = 26
    
# spliting data into samples
x_train, y_train = split_series(train_df.values,n_past, n_future)
x_train = x_train.reshape((x_train.shape[0], x_train.shape[1], n_features))
y_train = y_train.reshape((y_train.shape[0], y_train.shape[1], n_features))

x_test, y_test = split_series(test_df.values,n_past, n_future)
x_test = x_test.reshape((x_test.shape[0], x_test.shape[1],n_features))
y_test = y_test.reshape((y_test.shape[0], y_test.shape[1], n_features))
```

```python
# E1D1: Sequence to Sequence Model with one encoder layer and one decoder layer
# n_features ==> no of features at each timestep in the data.

encoder_inputs = tf.keras.layers.Input(shape=(n_past, n_features))
encoder_l1 = tf.keras.layers.LSTM(100, return_state=True)
encoder_outputs1 = encoder_l1(encoder_inputs)

encoder_states1 = encoder_outputs1[1:]

decoder_inputs = tf.keras.layers.RepeatVector(n_future)(encoder_outputs1[0])

decoder_l1 = tf.keras.layers.LSTM(100, return_sequences=True)(decoder_inputs,initial_state = encoder_states1)
decoder_outputs1 = tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(n_features))(decoder_l1)

model_e1d1 = tf.keras.models.Model(encoder_inputs,decoder_outputs1)

model_e1d1.summary()
```

```python
# E2D2: Sequence to Sequence Model with two encoder layers and two decoder layers
# n_features ==> no of features at each timestep in the data.

encoder_inputs = tf.keras.layers.Input(shape=(n_past, n_features))
encoder_l1 = tf.keras.layers.LSTM(100,return_sequences = True, return_state=True)
encoder_outputs1 = encoder_l1(encoder_inputs)
encoder_states1 = encoder_outputs1[1:]
encoder_l2 = tf.keras.layers.LSTM(100, return_state=True)
encoder_outputs2 = encoder_l2(encoder_outputs1[0])
encoder_states2 = encoder_outputs2[1:]

decoder_inputs = tf.keras.layers.RepeatVector(n_future)(encoder_outputs2[0])

decoder_l1 = tf.keras.layers.LSTM(100, return_sequences=True)(decoder_inputs,initial_state = encoder_states1)
decoder_l2 = tf.keras.layers.LSTM(100, return_sequences=True)(decoder_l1,initial_state = encoder_states2)
decoder_outputs2 = tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(n_features))(decoder_l2)

model_e2d2 = tf.keras.models.Model(encoder_inputs,decoder_outputs2)

model_e2d2.summary()
```

```python
reduce_lr = tf.keras.callbacks.LearningRateScheduler(lambda x: 1e-3 * 0.90 ** x)

model_e1d1.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.Huber(), metrics=['accuracy'])
history_e1d1=model_e1d1.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test),batch_size=32,verbose=1,callbacks=[reduce_lr])

model_e2d2.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.Huber(), metrics=['accuracy'])
history_e2d2=model_e2d2.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test),batch_size=32,verbose=1,callbacks=[reduce_lr])
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 4), tight_layout = True)

axes[0].plot(history_e1d1.history['loss'], label='loss')
axes[0].plot(history_e1d1.history['val_loss'], label='val_loss')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Error')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history_e2d2.history['loss'], label='loss')
axes[1].plot(history_e2d2.history['val_loss'], label='val_loss')
axes[1].set_xlabel('Epoch')
axes[1].legend()
axes[1].grid(True)
```

```python
test_results1, accuracy1 = model_e1d1.evaluate(x_test, y_test, verbose=False)
test_results2, accuracy2 = model_e2d2.evaluate(x_test, y_test, verbose=False)
print(test_results1, accuracy1)
print(test_results2, accuracy2)
```
