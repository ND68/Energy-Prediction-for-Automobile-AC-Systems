# 06/22 Data Analysis (Timestamps)

```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Data/0620-/CSVLog_20230622_110214.csv',comment="#") 
pd.set_option('display.max_columns', None) #shows all the columns

df = df.loc[:, (df != 0).any(axis=0)] #removes all columns that have only zeros in them
df.drop(index = [0,1,2,3,4,5], inplace=True)  #removes initial rows that don't contain any data
df.reset_index(drop = True, inplace=True)
df.columns = df.columns.str.lstrip(' ') #removes a random space char in front of column names

df.head()
```

```python
starttimedf = pd.read_csv('./Data/0620-/CSVLog_20230622_110214.csv', nrows = 0)
starttime = starttimedf.columns[0][14:]
starttime = pd.to_datetime(starttime).strftime('%s.%f')[:-3]
starttime = float(starttime)
starttime
```

```python
df.insert(loc=0, column='UnixTime', value = (df['Time (sec)'] + starttime)) #Epoch time
df.insert(loc=0, column='Datetime', value = (df['Time (sec)'] + starttime)) 
df['Datetime'] = pd.to_datetime(df['Datetime'], unit='s', utc = True).map(
    lambda x: x.tz_convert('America/Los_Angeles'))
df['Datetime'] = df['Datetime'].dt.round('ms')
#converts to datetime, rounds, timezones
df.head()
```

```python
df.set_index('Datetime', inplace = True)
df.head()
```

### Resampling ###

```python
df_10min = df.resample('10min').mean().round(3).reset_index()

df_5min = df.resample('5min').mean().round(3).reset_index()

df_1min = df.resample('1min').mean().round(3).reset_index()
df_1min.head(3)

# look into variable aggregates
```

```python
fig, axes = plt.subplots(2, 4, figsize = (18, 10), tight_layout = True)

# Time Vs A/C Energy Consumption
axes[0,0].plot(df[['Time (sec)']], df[['Aircon Consumption Power (kW)']])
axes[0,0].set(ylabel='Aircon Consumption Power (kW)')

axes[0,1].plot(df_1min[['Time (sec)']], df_1min[['Aircon Consumption Power (kW)']], color = 'lightcoral')
axes[0,1].set(ylabel='Aircon Consumption Power (kW)(1mins)')

axes[0,2].plot(df_5min[['Time (sec)']], df_5min[['Aircon Consumption Power (kW)']], color = 'sandybrown')
axes[0,2].set(ylabel='Aircon Consumption Power (kW)(5mins)')

axes[0,3].plot(df_10min[['Time (sec)']], df_10min[['Aircon Consumption Power (kW)']], color = 'lightgreen')
axes[0,3].set(ylabel='Aircon Consumption Power (kW)(10mins)')

# Time Vs Compressor Target Speed 
axes[1,0].plot(df[['Time (sec)']], df[['Compressor Target Speed (RPM)']])
axes[1,0].set(ylabel='Compressor Target Speed (RPM)')

axes[1,1].plot(df_1min[['Time (sec)']], df_1min[['Compressor Target Speed (RPM)']], color = 'lightcoral')
axes[1,1].set(ylabel='Compressor Target Speed (RPM)(1mins)')

axes[1,2].plot(df_5min[['Time (sec)']], df_5min[['Compressor Target Speed (RPM)']], color = 'sandybrown')
axes[1,2].set(ylabel='Compressor Target Speed (RPM)(5mins)')

axes[1,3].plot(df_10min[['Time (sec)']], df_10min[['Compressor Target Speed (RPM)']], color = 'lightgreen')
axes[1,3].set(ylabel='Compressor Target Speed (RPM)(10mins)')

fig.suptitle('Time vs A/C Power Consumption and Compressor Target RPM', fontsize = 20)
for ax in axes.flat:
    ax.set(xlabel='Time (sec)')
```
